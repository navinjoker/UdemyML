```python
# Import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Import data
bike_days = pd.read_csv('data/day.csv')
bike_hours = pd.read_csv('data/hour.csv')
```

```python
# Quick explain of the bike share data by day
print bike_days.describe()
print bike_days.head(5)
```

```python
## Looks like given the data set, we should use `instant` as the index, or dteday
bike_days_in = bike_days.set_index('instant')
bike_hours_in = bike_hours.set_index('instant')
bike_days_dt = bike_days
bike_days_dt['dteday'] = pd.to_datetime(bike_days_dt['dteday'])
bike_days_dt = bike_days_dt.set_index('dteday')
```

```python
## Observing relationship between date and temperature change
## temp = Celcius normalized to 0 to 1
print plt.plot(bike_days_dt.index, bike_days_dt['temp'], 'o')
```

```python
## Note that based on mean ~ median being relatively the same, and most of the data fits within bounds 5% and 95%
# we may expect a normal distribution of bike usage.
print bike_days['cnt'].describe()
print bike_days['cnt'].quantile(.05), bike_days['cnt'].quantile(.95)
print plt.hist(bike_days['cnt'], bins=10)
```

```python
# but by resizing the bins, we see a slightly different and wide shape
print plt.hist(bike_days['cnt'], bins=30)
```

```python
# So how about we observe the relationship between bike usage and the temperature outside?
# Is there an interesting story here?
print plt.plot(bike_days['cnt'], bike_days['temp'], 'o')
```

```python
# Here, compare bike usage to other weather effects:
# feeling temp (atemp), hum, windspeed, weather situation (weathersit)
print plt.plot(bike_days['cnt'], bike_days['atemp'], 'o')
```

```python
# How can we clean up the data outliers using holiday, workingday, weekday?
# Starting point: remove holidays. What else can we do?
bike_days_no_holiday = bike_days[bike_days['holiday'] == 0]
print plt.plot(bike_days_no_holiday['cnt'], bike_days_no_holiday['temp'], 'o')
```
