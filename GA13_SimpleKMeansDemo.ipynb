```python
plt.jet() # set a color map. 
import sklearn.datasets as datasets
chosencenters = [[1, 1], [-0.5, 0], [1, -1]]
cc = np.asarray(chosencenters)



###### Use Chosen Centers or Have Program Choose

X, Y = datasets.make_blobs(centers=chosencenters, cluster_std=(.5,.5)) # , random_state=0)
plt.scatter(cc[:,0], cc[:,1], s=100, c='black')

# Alternatively, comment above two linesand have program pick:
#X, Y = datasets.make_blobs(centers=3, cluster_std=(.5,.5)) # , random_state=0)

plt.scatter(X[:,0], X[:,1],c=Y);
```

As always, we first *plot* the data to get a feeling of what we're dealing with:

Normally, you do not know the information in `Y`, however.

You could try to recover it from the data alone.

This is what the kMeans algorithm does. 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(3, random_state=8)
Y_hat = kmeans.fit(X).labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat);
```



The means in $\mu$ can be seen as *representatives* of their respective cluster.

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c='black')
#print mu
```
